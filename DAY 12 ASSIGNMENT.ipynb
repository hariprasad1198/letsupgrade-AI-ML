# NON-PARAMETRIC TESTS 

# Problem Statement 1

H0: There is no significant difference in the employee's age between Attrition(Yes) and Attrition(No) ie Null hypothesis

H1: There is significant difference in the employee's age between Attrition(Yes) and Attrition(No) ie Alternate hypothesis

```python
from scipy.stats import mannwhitneyu
stats1, p1 = mannwhitneyu(d3.Age,d4.Age)
print("Critical value:",stats1)
print("p value:",p1)
```

# Conclusion

Clearly, the p value < 0.05 so null hypothesis is rejected and alternate hypothesis is accepted i.e There is significant differnce in the employee's age between Attrition(Yes)and Attrition(No).

```python
from scipy.stats import chi2_contingency
chitable1=pd.crosstab(d2.Attrition,d2.BusinessTravel)
chitable1
```

```python
stats13,p13,dof,expeted=chi2_contingency(chitable1)
print("Critical value:",stats13)
print("p value:",p13)
```
