**Observations from the correlation matrix-**

**1)** In the above correlation matrix fields like **"StandardHours"** and **"EmployeeCount"** don't have coefficient of correlation as the data entered in these fields have a constant value for all records in the table.

**2)** Many of the columns are not included in this correlation matrix as these column represent categorical variables which are first needed to be converted into integer type using label encoding. 

```python
#encodes if 'Yes'=1,'No'=0 for Attrition column
from sklearn.preprocessing import LabelEncoder
encoder1 = LabelEncoder()
df['Attrition'] = encoder1.fit_transform(df['Attrition'])

#encodes 'Travel_Rarely'=2,'Travel_Frequently'=1,'Non_Travel'=0 for BusinessTravel column
from sklearn.preprocessing import LabelEncoder
encoder2 = LabelEncoder()
df['BusinessTravel'] = encoder2.fit_transform(df['BusinessTravel'])

#encodes 'Sales'=2,'Research & Development'=1,'Human Resources'=0 for Department column
from sklearn.preprocessing import LabelEncoder
encoder3 = LabelEncoder()
df['Department'] = encoder3.fit_transform(df['Department'])

#encodes 'Technical Degree'=5,'Other'=4,'Medical'=3,'Marketing'=2,'Life Sciences'=1,'Human Resources'=0 for EducationField column
from sklearn.preprocessing import LabelEncoder
encoder4 = LabelEncoder()
df['EducationField'] = encoder4.fit_transform(df['EducationField'])

#encodes 'Male'=1,'Female'=0 for Gender column
from sklearn.preprocessing import LabelEncoder
encoder5 = LabelEncoder()
df['Gender'] = encoder5.fit_transform(df['Gender'])

#encodes 'Sales Representative'=8,'Sales Executive'=7,'Research Scientist'=6,'Research Director'=5,'Manufacturing Director'=4,'Manager'=3,'Laboratory Technician'=2,'Human Resources'=1,'Healthcare Representative'=0 in JobRole column
from sklearn.preprocessing import LabelEncoder
encoder6 = LabelEncoder()
df['JobRole'] = encoder6.fit_transform(df['JobRole'])

#encodes 'Single'=2,'Married'=1,'Divorced'=0 in MaritalStatus column
from sklearn.preprocessing import LabelEncoder
encoder7 = LabelEncoder()
df['MaritalStatus'] = encoder7.fit_transform(df['MaritalStatus'])

from sklearn.preprocessing import LabelEncoder
encoder8 = LabelEncoder()
df['Over18'] = encoder8.fit_transform(df['Over18'])
```

**Plotting heatmap**

```python
fig, ax = plt.subplots(figsize=(20,20))
sb.heatmap(df.corr(), ax=ax, annot=True, vmin=-1, vmax=1, center=0, cmap='coolwarm')
```

**Following conclusions can be drawn out-**

**1)** It can be clearly seen from the above heatmap based on the coefficient of correlation that **Attrition has high positive correlation with Marital status** of the employee, a more in depth analysis is required to figure out how is varies with single,married or divorced employee.

**2)** **Attrition has high negative correlation with Total working years and Age of an employee,and his/her Years with current manager and Years at company** i.e. one can conclude that Attrition tends to decreases with increase in Age,Total working years of an individual,his/her Years with current manager and at the company.   

**3)** Rest all fields or columns have negligible or zero correlation with Attrition.
